```python
# 크롬으로 웹 들어갈 것을 선택한것 크롬 없을시 엣지로 해도 가능
driver = webdriver.Chrome()
```

```python
# 지역별로 바꿔가면서 해야함
# 파일 빼는 것까진 안해놨으니 블로그 크롤링 파일보고 참고 해서 파일로 정리해서 빼주시면 됩니다
# 지금 보낸 파일은 전남 지역 크롤링한 것 예시
# 전남, 광주, 경남, 부산 지역 크롤링해서 파일로
# csv파일로 만들 것
driver.get("http://ticket.interpark.com/TiKi/Special/TPRegionReserve.asp?Region=42040&RegionName=%C0%FC%B3%B2")
```

```python
# 2. 영상의 제목만 크롤링
soup = bs(driver.page_source, 'lxml')
```

```python
date = soup.select('.date')
```

```python
place = soup.select('.place')
```

```python
img = soup.select('.name')
```

```python
soup.find('dd',class_="name")
```

```python
a = soup.find('dd',class_="name").find_all('a')
```

```python
title = soup.select('a')
```

```python
# 타이틀만 가져오는 슬라이싱 갯수가 달라지면
# 슬라이싱 시작점과 끝나는 점이 달라지니 지역이나 기간을 달리 할 경우 다시 확인하기
title_all1 = title[427:-14]
```

```python
title_all = []
for i in title_all1:
    title_all.append(i.text.strip())
title_all
```

```python
date_all = []
for i in date:
    date_all.append(i.text.strip())
```

```python
# 기간이 알맞게 들어갔는지 확인하는 용도
# 쓸데없는 글씨는 제거해놓은 열
while '기간' in date_all:
    date_all.remove('기간')
date_all
```

```python
place_all = []
for i in place:
    place_all.append(i.text.strip())
place_all2 = place_all[1:]
```

```python
while '장소' in place_all2:
    place_all2.remove('장소')
```

```python
title = soup.select('.name')
```

```python
title_all1 = title[1:]
```

```python
title_all = []
for i in title_all1:
    title_all.append(i.text.strip())
title_all = title_all[:-1]
```

```python
# 중간에 매치가 안맞게 하는 글씨 제거
while '상품명' in title_all:
    title_all.remove('상품명')
```

```python
a = content[0].find_element(By.TAG_NAME, "a").get_attribute("href")

urls = list(map(
    lambda item : item.find_element(By.TAG_NAME, "a").get_attribute("href"),
    content
))
```

```python
urls
```

```python
img = content[0].find_element(By.TAG_NAME, "img").get_attribute("src")

imgs = list(map(
    lambda item : item.find_element(By.TAG_NAME, "img").get_attribute("src"),
    content
))
```

```python
date_last = []
for i in date:
    date_last.append(i.text[13:])
```

```python
date_first = []
for i in date:
    date_first.append(i.text[:-13])
```

```python
# 중간에 매치가 안맞게 하는 글씨 제거
# 시작 날짜
while '' in date_first:
    date_first.remove('')
```

```python
# 중간에 매치가 안맞게 하는 글씨 제거
# 끝나는 날짜
while '' in date_last:
    date_last.remove('')
```
