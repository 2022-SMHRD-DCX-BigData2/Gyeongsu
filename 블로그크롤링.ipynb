```python
import time

section = driver.find_element(By.TAG_NAME,'section')

body = driver.find_element(By.TAG_NAME, 'body')

totalCount = 60 # 가져올 대상 개수


retry = 0
retry_limit = 10 # 페이지 무한 로딩을 방지하기 위해 재시도 횟수 설정

prev_item_count = -1 # 직전 루프에 원소 갯수

while True:
    items = driver.find_elements(By.CSS_SELECTOR, 'a.api_txt_lines.total_tit._cross_trigger')
    count = len(items)

    if prev_item_count == count:
        retry += 1

        if retry >= retry_limit:
            break
        
        body.send_keys(Keys.END)
        time.sleep(0.1)
        continue

    retry = 0

    prev_item_count = count

    if count >= totalCount:
        break

    body.send_keys(Keys.END)
    time.sleep(0.1)
```
